```python
# user defined function to generate statistics related to Categorical Variables
def categorical_stats(col):
    cat_df = bike.groupby(col)['cnt'].agg(['sum', 'mean','count']).sort_values('sum',ascending = False)
    cat_df['sum_perc']=cat_df['sum']/bike.cnt.sum()*100
    cat_df['count_perc']=cat_df['count']/bike.cnt.count()*100
    return round(cat_df,2)
```

```python
# user defined function to generate plots related to Categorical Variables
def categorical_plot(col,x,y):
    plt.figure(figsize = (x,y))
    plt.subplot(1,2,1)
    sns.barplot(col,'cnt',data=bike)
    plt.subplot(1,2,2)
    sns.barplot(col,'cnt',data=bike, hue='yr',palette='Paired')
    plt.legend(labels=['2018', '2019'])
    return
```

```python
# Variance Inflation Factor(VIF)
def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# dataframe containing all the variables and their respective VIFs
calculateVIF(X_train_rfe)
```

```python
# Add a constant
X_train_lm1 = sm.add_constant(X_train_rfe)

# Create a first fitted model
lr1 = sm.OLS(y_train, X_train_lm1).fit()

# Print a summary of the linear regression model obtained
print(lr1.summary())
```

```python
# The above refined model has low Multicollinearity between the predictors and the p-values for all the predictors seems to be significant
# This can be taken as our final model
# as our coefficients are not equal to zero, We can reject the null hypothesis
#Oue model looks significant as it has higher F value (253) and Prob (F-statistic) is 3.13e-188. The higher the F-Statistics, the more significant the Model is.
```

```python
# The equation of best fitted model as per lr6:

# cnt=0.075009 + (0.233139 x yr) + (0.056117 x workingday) + (0.549892 x temp) + (-0.155203 x windspeed) + (0.088621 x W2_Summer) + (0.130655 x W4_Winter) + (0.097365 x Sep) + (-0.287090 x Light_rainsnow) + (-0.080022 x Misty) + (0.067500 x Saturday)
```

```python
# Interpretation of the above equation

# 1. const : The Constant value of ‘0.075009’ indicates that, in the absence of all other predictor variables 
#     (i.e. when x1,x2...xn =0), The bike rental can still increase by 0.075009 units

    
# 2. yr : A coefficient value of ‘0.233139’ indicates that a unit increase in yr variable, increases the bike hire numbers by 0.233139 units
 
# Similarly, for other variables' cofficient interpretation remains same
```

```python
#Adjusted R2 value
# n is number of rows in test dataset
n = X_test.shape[0]

# Number of features (predictors, p) is the shape along axis 1
p = X_test.shape[1]

# We find the Adjusted R-squared using the formula
adjusted_r2 = round(1-(1-r2)*(n-1)/(n-p-1),4)
adjusted_r2
```
